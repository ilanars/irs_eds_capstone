### Create Capstone Project Structure

```python
# Create your project structure inside irs_eds_capstone

create_project_structure <- function(irs_eds_capstone) {
  "Create a well-organized directory structure for your research project"
  
  # Check that main project folder already exists
  if (!dir.exists(irs_eds_capstone)) {
    message("Main project directory does not exist â€” creating it now...")
    dir.create(irs_eds_capstone, recursive = TRUE)
  }
    
  # Create data and scripts directories
  data_path <- file.path(irs_eds_capstone, "data")
  scripts_path <- file.path(irs_eds_capstone, "scripts")
  setup_path <- file.path(irs_eds_capstone,"setup")
  
  dir.create(data_path, showWarnings = FALSE)
  dir.create(scripts_path, showWarnings = FALSE)
  dir.create(setup_path, showWarnings = FALSE)
    
  # Define our 6-folder system
  data_folders <- c(
    "raw_data",
    "processed_data",
    "analysis",
    "metadata",
    "outputs",
    "archive"
  )
  
  # Create each data subfolder
  for (folder in data_folders) {
    folder_path <- file.path(data_path, folder)
    dir.create(folder_path, showWarnings = FALSE)
    
    # Add a README to explain each folder's purpose
    readme_path <- file.path(folder_path, "README.txt")
    
    # Write helpful descriptions
    if (grepl("raw_data", folder)) {
      writeLines(
        c("IMPORTANT: Never modify files in this folder!",
          "This is your data vault - keep originals safe here."),
        readme_path
      )
    } else if (grepl("processed", folder)) {
      writeLines(
        c("Place cleaned and processed data here.",
          "Always document what processing was done."),
        readme_path
      )
    } else if (grepl("analysis", folder)) {
      writeLines(
        c("Extra analysis docs / notebooks go here.",
          "Organize by analysis type or date."),
        readme_path
      )
    } else if (grepl("metadata", folder)) {
      writeLines(
        c("Documentation about your data.",
          "Include data dictionaries, collection methods, etc."),
        readme_path
      )
    } else if (grepl("outputs", folder)) {
      writeLines(
        c("Final results, figures, and reports.",
          "Ready to share with others!"),
        readme_path
      )
    } else if (grepl("archive", folder)) {
      writeLines(
        c("Old versions and superseded files.",
          "Always date your archived files."),
        readme_path
      )
    }
  }
  
  # Add README for scripts folder
  scripts_readme <- file.path(scripts_path, "README.txt")
  writeLines(
    c("Version-controlled scripts and code.",
      "Use git to track changes to your analysis code."),
    scripts_readme
  )
 
    
  # Add README for setup folder
  setup_readme <- file.path(setup_path, "README.txt")
  writeLines(
    c("Folder to hold your setup script and project psuedocode"),
    setup_readme
  )
    
  # List the created structure
  cat("\nCreated folders:\n")
  all_dirs <- list.dirs(irs_eds_capstone, recursive = TRUE, full.names = FALSE)
  for (dir in all_dirs[all_dirs != ""]) {
    cat(sprintf("  ðŸ“ %s\n", dir))
  }
  
  return(invisible(TRUE))
}

# Create an example project
create_project_structure("irs_eds_capstone")
```
